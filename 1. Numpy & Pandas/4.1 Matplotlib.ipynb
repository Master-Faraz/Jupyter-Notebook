```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.scatter(x,y,color='red',s=50)  #  S is for marker size
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y,color='red',linestyle=':',marker='*',markersize=10,linewidth=2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
```
