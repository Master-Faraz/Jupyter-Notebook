# Training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.data , df.target , test_size = 0.2 )
```

```python
from sklearn.linear_model import LogisticRegression
```

### max_iter is used to resolve the iteration overflow issue

```python
model = LogisticRegression(max_iter=1000000)       
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_test , y_test)
```

# Confusion matrix and its Visualization

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

**This is the visualization of confusion matrix**

As we can see above our model fails at many points like<br>
in 2,2 from left-bottom side <br>
