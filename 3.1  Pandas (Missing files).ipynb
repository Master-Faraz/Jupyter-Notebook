#  To Handle NaN  values we use these methods

### 1. Filling all values to a number

```python
new_df = df.fillna(0)
new_df
```

### 2. filling specific values to each columns

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

### 3.  Filling with data of previous events so that our data doesn't looks bad

```python
new_df = df.fillna(method='ffill')       #   Forward Fill
new_df
```

```python
new_df = df.fillna(method='bfill')       #   Backward Fill
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df  #  Here it is columns so it will copy columns data
```

```python
new_df = df.fillna(method="bfill", axis="index") 
new_df
```

### 4. Filling data with limit of copy data

```python
new_df = df.fillna(method='ffill',limit=1)       #   Forward Fill with limit 1
new_df
```

###  5.   Predicting data by taking average 

```python
new_df = df.interpolate()
new_df
```

### ***6. It predicts according to date and time   See Second row it predicts according to time***

```python
new_df = df.interpolate(method="time")   
new_df
```

### 7.  It Drops all NaN   Values

```python
new_df = df.dropna()
new_df
```

**Drop if all NaN**

```python
new_df = df.dropna(how='all')
new_df
```

**IF atleast 1 non NaN value then preserve the raw**

```python
new_df = df.dropna(thresh=1)
new_df
```

### 8. For filling Missing dates

#  Replace Method : - 


```python
import numpy as np
df = pd.read_csv("/home/faraz/Documents/Python Projects/weather_data.csv")
df
```

**When we get data from online source sometimes we get data like -99999 or any other garbage value so we use replace method to handle this issue**

```python
new_df = df.replace(-99999, value=np.NaN) 
new_df
```

```python
new_df = df.replace(to_replace=[-99999,-88888], value=0)    #  When we have more than one garbage value
new_df
```

**Replacing Values column by column **

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df
```

**Replacement based on columns**

```python
new_df = df.replace({
        -99999: np.nan,
        '0': 'Sunny',
    })
new_df

```

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.  
# REGEX

new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df

```
